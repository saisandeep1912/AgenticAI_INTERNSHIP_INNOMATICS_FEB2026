# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyser(likes_list):
    total_likes=0
    for i in likes_list:
        total_likes+=i
    if total_likes>1000:
        status='Viral Post'
    else:
        status="Normal Engagement"
    print("Total Likes:",total_likes)
    print("Post Status:",status)

    
likes_data=[200,300,150,250,150]
analyser(likes_data)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall_data,required_level):
    total=0
    for i in rainfall_data:
        total+=i
    average=total/len(rainfall_data)
    if average>required_level:
        status='Adequate Rainfall'
    else:
        status='Inadequate Rainfall'
    print('Average Rainfall: ',average)
    print('Rainfall Status: ',status)
rainfall_data=[70,75,68,80,67]
required_level=70
rainfall_checker(rainfall_data,required_level)
```

# Problem 4: Social Media – Duplicate Account Detection

```python
def detect_dup_acc(usernames):
    unq_user=set()
    dup_found=False
    for i in usernames:
        if i in unq_user:
            dup_found=True
            break
        else:
            unq_user.add(i)
    if dup_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print('Duplicate Accounts Found: No')
usernames=["john123", "sai_01", "alex99", "john123"]
detect_dup_acc(usernames)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):
    premium_crops=[]
    for i in prices:
        if i >2000:
          premium_crops.append(i)
    print("Premium Crops:",premium_crops)
prices=[1500,2500,1000,3200]
premium_crop_filter(prices)
```

# Problem 7: System Monitoring – Application Health Checker


```python
def application_health_checker(error_count):
    if error_count ==0:
        status='Healthy'
    elif error_count<=5:
        status='Minor Issues'
    elif error_count>5:
        status='Critical Issues'
    print('Error Count:',error_count)
    print('System Status: ',status)
application_health_checker(7)
    
```

# Problem 8: Banking – Daily Transaction Limit Checker


```python
def transaction_limit_checker(transactions):
    total=0
    for i in transactions:
        total+=i
    if total <= 50000:
        status='Approved'
    else:
        status='Rejected'
    print("Transaction Amount:",total)
    print("Transaction Status:",status)
transactions=[10000,25000,15000,10000]
transaction_limit_checker(transactions)
```

# Problem 9: E-Learning – Student Attendance Eligibility System


```python
def attendance_eligibility(attendance_list):
    total_days=len(attendance_list)
    present_days=0
    for i in attendance_list:
        if i=='P':
            present_days+=1
    percentage=(present_days/total_days)*100
    if percentage>=75:
        status='Eligible'
    else:
        status='Not Eligible'
    print('Attendance Percentage:',percentage)
    print('Exam Eligibility:',status)
attendance_eligibility(attendance_list=['P','P','A','P','P'])
```

# Problem 10: Smart Electricity Bill Analyzer


```python
def electric_bill_analyser(units_consumed):
    if units_consumed<=100:
        cost=units_consumed*3
    elif 100 < units_consumed <=200:
        n=units_consumed-100
        cost= 300+(n*5)
    elif units_consumed >200:
        x=units_consumed-200
        cost=800+(x*7)
    if cost<500:
        status='Low Usage'
    elif 500 <=cost<=1500:
        status='Moderate Usage'
    elif cost>1500:
        status='High Usage'
    return cost,status
units_consumed=273
total_bill,usage_status=electric_bill_analyser(units_consumed)
print('Total_bill:',total_bill)
print("Usage_Status:",usage_status)
```
