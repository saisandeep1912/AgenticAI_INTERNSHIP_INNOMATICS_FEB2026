# Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking(capacity,logs):
    current_vehicle=0
    peak_usage=0

    for log in logs:
        if 'IN'== log:
            current_vehicle+=1
        elif 'OUT' == log:
            current_vehicle-=1
    if current_vehicle>peak_usage:
        peak_usage=current_vehicle
    if current_vehicle>capacity:
        status='Exceeded Capacity'
    else:
        status='Available'
    print('Currently Parked Vehicles:',current_vehicle)
    print("Peak Usage",peak_usage)
    print("Parking Status:",status)


capacity=50
logs=["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(capacity,logs)
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def eta(distance,traffic,weather):
    base_time=(distance/24)*60 #24kmh-average speed
    if traffic =='Low':
        traffic_delay=5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0

    if weather =='Clear':
        weather_delay=0
    elif weather =='Rainy':
        weather_delay=15
    elif weather =='Strom':
        weather_delay=25
    else:
        weather_delay=0
    final_time=base_time+traffic_delay+weather_delay
    print('Estimated Delivery Time:',final_time)

eta(8,"High","Rainy")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyse_seat_occ(total_seats,booked_seats):
    total_bookes=len(booked_seats)
    occupancy_percentage=(total_bookes/total_seats)*100

    if occupancy_percentage==100:
        status='Housefull'
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    elif occupancy_percentage >= 50:
        status = "Filling Fast"
    else:
        status = "Seats Available"
    if occupancy_percentage >= 90:
        suggestion = "Consider Opening Additional Show"
    else:
        suggestion = "No Additional Show Needed"
    print("Occupancy:", round(occupancy_percentage), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)

total_seats=200
booked_seats=[1]*150
analyse_seat_occ(total_seats,booked_seats)
```

# Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_reading):
    total=0
    count=0
    for reading in cpu_reading:
        total+=reading
        count+=1
    average_cpu=total/count
    
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", round(average_cpu), "%")
    print("Server Status:", status)
classify_server_load(cpu_reading=[45, 60, 70, 85, 90])
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def monitor_resource_usage(resource_usage):
    overused_resources = []
    threshold = 8
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)
    if len(overused_resources) > 0:
        alert='Yes'
    else:
        alert='No'
    print('Overused Resources:',",".join(overused_resources) if overused_resources else 'None')
    print('Energy Alert:',alert)

resource_usage = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}
monitor_resource_usage(resource_usage)
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, total_registrations):
    confirmed=0
    waitlisted=0
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
manage_registrations(capacity=100, total_registrations=105)
```
