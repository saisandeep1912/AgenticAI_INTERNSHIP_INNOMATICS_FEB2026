```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

highest_score = max(employees.values())
top_perf=[]
for i,j in employees.items():
    if j==highest_score:
        top_perf.append(i)
print("Top Performers Eligible for Bonus:",",".join(top_perf),"(Score:",highest_score,")")
```

```python
query = "Buy mobile phone buy phone online"
query = query.lower()
cleaned_query = ""
for char in query:
    if char.isalnum() or char.isspace():
        cleaned_query += char
words = cleaned_query.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
result = {}
for word, count in frequency.items():
    if count > 1:
        result[word] = count
print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for index in range(len(sensor_readings)):
    reading = sensor_readings[index]
    if reading % 2 == 0:
        valid_readings.append((index, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
threshold = average * 1.30
for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```
