# Problem 1: Mobile Recharge Validation System

```python
def recharge_validator(amount):
    valid_plan=[199,299,399,599]
    
    if amount <50:
        print("Recharge must be atleast 50")
        return False
    if amount not in valid_plan:
        print("Invalid plan selected.")
        print("Available plans:", valid_plan)
        return False
    print('Recharge Successful')
    return True
while True:
    amount=int(input("Enter the amount of Recharge:"))
    if recharge_validator(amount):
        break
```

# Problem 2: Inventory Reorder Alert System

```python
def check_stores(products):
    for product,quantity in products.items():
        if quantity<15:
            print(f"{product} Reorder Alert (Stock:{quantity})")
        else:
            print(f"{product} Stock Ok (Stock:{quantity})")
products={}
n = int(input("Enter number of products: "))
for i in range(n):
    name=input(f"Enter product number{i+1}:")
    quantities=int(input("Enter quantity:"))
    products[name]=quantities

check_stores(products)
```

# Problem 3: Student Result Processing System

```python
def process_result(marks):
    total=0
    for i in marks:
        total+=i
    average= total/len(marks)
    if average>=50:
        result="Pass"
    else:
        result='Fail'

    return average,result
marks=[70,60,50]
avg,res=process_result(marks)
print(f'Average Marks:{avg}, Result:{res}')
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12
    fare=base_fare+(per_km_rate*distance)
    if is_peak:
        fare=fare*1.25
    return fare
while True:
    distance=float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()
    if peak_input == "yes":
        is_peak = True
    else:
        is_peak = False
    total_fare = calculate_fare(distance, is_peak)
    print(f"Total Fare: ₹{total_fare}")
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator! ")
        break
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance):
    present_count = 0
    for status in attendance:
        if status == "P":
            present_count += 1
    total_days = len(attendance)
    percentage = (present_count / total_days) * 100
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"
    return percentage, result
attendance_list = ["P", "A", "P", "P", "A", "P"]
percent, eligibility = check_eligibility(attendance_list)
print("Attendance Percentage:", percent)
print("Status:", eligibility)
```

# Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character @#$)"
pwd = input("Enter password: ")
result = check_password(pwd)
print("Password Status:", result)
```
